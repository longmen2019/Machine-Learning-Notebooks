### Drawback of Ordinary Least Squares (OLS) Regression

While Ordinary Least Squares (OLS) Regression is a widely used method, it has some notable drawbacks:

- **Sensitivity to Outliers**: OLS is not robust, meaning it is highly sensitive to outliers. The presence of outliers can significantly skew the results, as OLS will attempt to minimize the overall cost by adjusting the regression line towards the outliers.

This sensitivity can lead to misleading conclusions and reduced accuracy in the presence of anomalous data points. Therefore, it's important to consider alternative robust regression methods or preprocessing steps to handle outliers effectively.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sequence of numbers from 1 to 100
x = np.linspace(1, 100, 100)

# Create a linear relationship with some noise
y = x * 2

# Introduce outliers in the data
y[10:30] = np.random.rand(20) * 120 + 100

# Reshape x and y to be 2D arrays, as required for certain machine learning models
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

# Plot the data to visualize the relationship and outliers
plt.scatter(x, y, color='blue', label='Data Points')
plt.title('Data with Outliers')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

We can observe some perturbations in the scatter plot. Next, we'll examine how Ordinary Least Squares (OLS) fits the data.


```python
from sklearn.linear_model import LinearRegression

# Fit the Linear Regression model
lr = LinearRegression().fit(x, y)

# Plot the data and the OLS regression line
plot(x, y, clf=lr, clf_name="OLS", color="green")
```

```python
from sklearn.linear_model import TheilSenRegressor

# Fit the Theil-Sen Regressor model
tr = TheilSenRegressor().fit(x, y)

# Plot the data and the Theil-Sen regression line
plot(x, y, clf=tr, clf_name="Theil-Sen Regressor", color="red")
```
