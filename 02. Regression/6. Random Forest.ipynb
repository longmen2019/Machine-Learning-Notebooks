### Decision Trees and Random Forests

**Decision Trees** tend to overfit the model. This means they can capture noise in the training data, leading to poor generalization on new, unseen data.

**Random Forests** address this issue by creating multiple decision trees using random subsets of the data and features. The final prediction is obtained by averaging the outcomes of all the trees. This process, known as **ensemble learning**, helps to:

- **Reduce Overfitting**: By averaging multiple trees, random forests smooth out the predictions, making the model more robust and less likely to overfit.
- **Improve Accuracy**: The aggregation of multiple trees leads to more accurate and reliable predictions.



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv('Position_Salaries.csv')

# Extract the independent variable (Position level) and dependent variable (Salary)
X = df.iloc[:, 1:2].values  # Using .values to convert to numpy array
Y = df.iloc[:, 2].values

# Display the first few rows of the dataframe
print(df.head())

# Optional: Plotting the data to visualize the relationship
plt.scatter(X, Y, color='blue')
plt.title('Position Level vs Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Assuming X and Y are already defined as in the previous code
# X = df.iloc[:, 1:2].values
# Y = df.iloc[:, 2].values

# Initialize and fit the Random Forest Regressor
regressor = RandomForestRegressor(n_estimators=1000, random_state=0)
regressor.fit(X, Y)

# Visualize the original data points
plt.scatter(X, Y, color='blue', label='Actual Data')

# Create a grid of values for X to plot a smooth curve
X_grid = np.arange(min(X), max(X), 0.01).reshape(-1, 1)

# Plot the Random Forest Regression results
plt.plot(X_grid, regressor.predict(X_grid), color='red', label='Model Prediction')

# Add titles and labels
plt.title('Random Forest Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend()

# Display the plot
plt.show()
```
