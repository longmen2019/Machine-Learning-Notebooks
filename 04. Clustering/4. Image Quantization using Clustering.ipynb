## Image Quantization

**Image Quantization** is a technique used to compress images by converting continuous values into discrete values. This is a form of lossy compression, which significantly reduces the file size.

### Using KMeans Clustering for Image Compression

We can use **KMeans clustering** to compress an image. The process involves:

1. **Converting RGB values**: The RGB values, which range from 0 to 255, are reduced to only 5 values.
2. **Transforming the image**: The image is converted from its original RGB format to one with only 5 colors.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure with a specified size
fig = plt.figure(figsize=(14, 10))

# Read the image from the file
image = plt.imread("one_pience.png")

# Set the title of the plot
plt.title("Original Image")

# Display the image
plt.imshow(image)

# Show the plot
plt.show()
```

```python
# The original image has three dimensions: one for each R, G, and B channel.
# We'll convert it into a 2D array by reducing one dimension.

# Get the dimensions of the image
x, y, z = image.shape

# Reshape the image to a 2D array where each row is a pixel and each column is a color channel (R, G, B)
image = image.reshape(x * y, z)

# Print the new shape of the image to verify the transformation
print(image.shape)
```

```python
from sklearn.cluster import KMeans

# Initialize the KMeans model with the desired number of clusters
clf = KMeans(n_clusters=5)

# Fit the KMeans model to the image data
clf.fit(image)
```

```python
# Retrieve the cluster centers (the RGB values of the 5 clusters)
centers = clf.cluster_centers_

# Retrieve the labels for each pixel (which cluster each pixel belongs to)
labels = clf.labels_

# Create a new figure with a specified size
fig = plt.figure(figsize=(14, 10))

# Set the title of the plot
plt.title("Compressed Image")

# Map each pixel to the RGB value of its cluster center and reshape to the original image dimensions
compressed_image = centers[labels].reshape(x, y, z)

# Display the compressed image
plt.imshow(compressed_image)

# Show the plot
plt.show()
```
