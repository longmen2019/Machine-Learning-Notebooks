```python
# Store the data into variables
df1 = pd.read_csv('dow_jones_industrial_average_news.csv')
df2 = pd.read_csv('dow_jones_industrial_average_stock.csv')
```

```python
# Merge the two datasets on the 'date' column using an inner join
merge = df1.merge(df2, how='inner', on='Date')

# Display the merged dataset
merge
```

```python
# Combine the top news headlines from the merged dataset
headlines = []
for row in range(0, len(merge.index)):
    headlines.append(' '.join(str(x) for x in merge.iloc[row, 2:27]))

# Print a sample of the combined headlines (first one)
print(headlines[0])
```

```python
import re

# Clean the data by removing unwanted characters from the headlines
clean_headlines = []
for i in range(len(headlines)):
    # Remove "b'" and 'b"' from the text
    clean_headlines.append(re.sub(r"b[(')]", '', headlines[i]))
    clean_headlines[i] = re.sub(r'b[(")]', '', clean_headlines[i])
    # Remove "\'" from the text
    clean_headlines[i] = re.sub(r"\\'", '', clean_headlines[i])

# Display the first cleaned headline as a sample
print(clean_headlines[0])
```

```python
# Add the cleaned headlines to the merged dataset as a new column
merge['Combined_News'] = clean_headlines

# Display the first entry of the new column to verify the addition
print(merge['Combined_News'][0])
```
